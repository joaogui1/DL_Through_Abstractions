```python
import numpy as np
import pandas as pd
```

```python
class Perceptron:
    def __init__(self, n):
        self.weights = np.random.uniform(-1, 1, n)
        self.bias = np.random.uniform(-1, 1)
    def feedforward(self, x):
        return np.sign(np.dot(self.weights, x) + self.bias)
    def update_weights(self, x, error):
        self.weights += error*x
        self.bias += error
    def train(self, X, T, epochs=500):
        for _ in range(epochs):
            sum_error = 0
            for x, t in zip(X, T):
                error = t - self.feedforward(x)
                self.update_weights(x, error)
                sum_error += np.abs(error)
            print(sum_error)
            if sum_error == 0.0:
                break
    
```

```python
df = pd.read_csv("linsep.csv", names=['x1', 'x2', 'y'])
```

```python
inputs = np.asarray([[x1, x2] for x1, x2 in zip(df['x1'], df['x2'])], dtype=np.float32)
outputs = df['y'].values
```

```python
net = Perceptron(2)
net.train(inputs, outputs)
```
